```python
import os
from pathlib import Path
from Replay import Replay
```

```python
base_path = Path('replay_files')
files = [file for file in os.listdir(base_path)]
print(files)
```

```python
replays = [Replay(base_path / file) for file in files]
print("Done")
```

```python
indentation = "  "
for replay in replays:
    replay.generate_player_stats()
    
    print(replay.filename)
    for player in replay.players_stats:
        print(indentation + player.name)
        print(indentation * 2 + "joined: " + str(player.join_times))
        print(indentation * 2 + "left: " + str(player.leave_times))
        print(indentation * 2 + "tagged times: " + str(player.tag_times))
        

        
```
